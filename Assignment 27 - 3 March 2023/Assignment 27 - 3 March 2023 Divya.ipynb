__Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.__

__Ans.__

Seaborn is a powerful Python library for data visualization built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. Scatter Plot:
   - Uses: Scatter plots are used to visualize the relationship between two continuous variables. They are useful for identifying patterns, clusters, or correlations between the variables.

2. Line Plot:
   - Uses: Line plots are typically used to visualize the trend or change in a variable over time. They are effective for showing the progression or comparison of data points in a series.

3. Bar Plot:
   - Uses: Bar plots are used to display and compare categorical data. They are useful for visualizing the distribution of a categorical variable or comparing the values of different categories.

4. Histogram:
   - Uses: Histograms are used to represent the distribution of a continuous variable. They provide a visual representation of the data's frequency or density, helping to identify patterns, peaks, or skewness.

5. Heatmap:
   - Uses: Heatmaps are effective for visualizing the correlation between variables in a dataset. They use color gradients to represent the magnitude of the relationship, making it easier to identify patterns or clusters.

These are just a few examples of the wide range of plots available in Seaborn. The library offers many other types of plots, such as box plots, violin plots, pair plots, and more, each serving specific visualization needs depending on the nature of the data and the analysis goals.

__Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.__

__Note: timepoint, signal, event, and region are columns in the fmri dataset.__

__Ans.__

```python
#importing the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")
```

```python
fmri_data.shape
```

```python
#Display first 10 rows of dataset
fmri_data.head(10)
```

```python
fmri_data['event'].value_counts()
```

```python
fmri_data['region'].value_counts()
```

```python
# Create the line plot using Seaborn
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")
plt.title('Signal over Time for Different Events and Regions')
# Display the plot
plt.show()
```

__Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.__

__Note: pclass, age, and fare are columns in the titanic dataset.___

__Ans.__

```python
#importing the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data.shape
```

```python
titanic_data.head()
```

```python
# Create the first box plot for age
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Create the second box plot for fare
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
# Display the plots
plt.show()
```

__Q6. Use the "flights" dataset from seaborn to plot a heatmap.__

__Ans.__

```python
#importing the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the flights dataset
flights_data = sns.load_dataset("flights")
```

```python
flights_data.shape
```

```python
flights_data.head()
```

```python
# Create the flights_matrix using pivot
flights_matrix = flights_data.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

```python
# Create the heatmap
sns.heatmap(data=flights_matrix, cmap="YlGnBu", annot=True, fmt="d")
# Add labels and title
plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Number of Passengers (1000s) by Year and Month")
# Display the plot
plt.show()
```
