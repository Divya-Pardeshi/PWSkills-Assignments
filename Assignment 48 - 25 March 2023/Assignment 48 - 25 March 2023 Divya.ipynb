Impact on Analysis:
-    Skewing Descriptive Statistics: Outliers can heavily influence summary statistics like the mean and standard deviation. The mean can be skewed by extreme values, making it an unreliable measure of central tendency. Similarly, the standard deviation might be significantly affected, misleading the assessment of data variability.

-    Altering Data Distribution Perception: Outliers might give the impression of a different distribution in the data than what it actually represents. For instance, a dataset might seem normally distributed, but outliers can distort this perception.

-    Affecting Relationships in Analysis: In analyses involving regression or correlation, outliers can significantly impact the relationships observed. They might strongly influence the results, potentially misleading the interpretation of correlations or regression coefficients.

-    Statistical Assumptions: Outliers can violate assumptions of some statistical methods. For example, parametric statistical tests might assume normality and homogeneity of variance, which outliers can disrupt, leading to inaccurate results.

**Q13.How does the type of app affect its price? Create a bar chart to compare average prices by app type.**

### Observation
- Developing a Finace, Lifestyle, medical app have highest prices in comparision to others.

```python
df['Price'] = df['Price'].str.extract('(\d+)')
df['Price'] = df['Price'].astype(float)
```

```python
average_prices = df.groupby('Category')['Price'].mean().reset_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))

plt.bar(average_prices['Category'],  average_prices['Price'], color='skyblue')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')

# Rotate x-labels for better readability (if needed)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Group by 'Category' and sum the 'Prices' for each app
Price_stats = df.groupby('Category')['Price'].mean()

# To visualize the top 10 most Priced apps:
top_10_Price = Price_stats.nlargest(10)

print(top_10_Price)
```

```python
plt.figure(figsize=(8, 6))
top_10_Price.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Price')
plt.title('Top 10 Installed Apps')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Q14.What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.**


```python
df['Installs'].unique()
```

```python
replace = ['+',',','$']
for x in  replace:
    df['Installs'] = df['Installs'].str.replace(x,'')
    
df = df[df['Installs'] !='Free']

df['Installs'] = df['Installs'].astype(int)

df['Installs'].unique()
```

```python
# Group by 'App Name' and sum the 'No. of Installs' for each app
install_stats = df.groupby('App')['Installs'].sum()

# To visualize the top 10 most installed apps:
top_10_installed = install_stats.nlargest(10)


print(top_10_installed)
```

```python
plt.figure(figsize=(8, 6))
top_10_installed.plot(kind='bar', color='skyblue')
plt.xlabel('App Name')
plt.ylabel('Total Installs')
plt.title('Top 10 Installed Apps')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
